```python
#identifying columns having missing values nad filling it with mean and returning the new dataframe
listNaCols = [miss for miss in dfWine.columns if dfWine[miss].isna().sum()>0] 
listNaCols 
for i in listNaCols: 
    dfWine[i].fillna(dfWine[i].mean(), inplace=True)
```
